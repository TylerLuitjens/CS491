```
import matplotlib.pylab as plt

def create_SIR_model(R0, beta, gamma):
  N = 10000 #as defined by prof tessmer
  S = N - 1
  I = 1
  R = 0
  
  #S,I,R
  S_data = []
  I_data = []
  R_data = []
  
  
  #fill in model
  for t in range (0,100):
    S_new = S - ((beta * S * I)/N)
    I_new = I + ((beta * S * I)/N) - (I * gamma)
    R_new = R + gamma * I
    
    S = S_new
    I = I_new
    R = R_new
    
    S_data.append(S)
    I_data.append(I)
    R_data.append(R)
  
  graph = plt.figure()
  
  S_line = plt.plot(S_data,label='S')
  I_line = plt.plot(I_data,label='I')
  R_line = plt.plot(R_data,label='R')
  
  plt.title('SIR Model')
  plt.legend()
  plt.xlabel('Time T')
  plt.ylabel('Population N')
  plt.show()
```
